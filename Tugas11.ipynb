```python
import datetime
import random
import string

perpustakaan_template = {
    'judul_buku':'judul',
    'penulis': 'penulis',
    'tahun_terbit': 1111,
    'tanggal_terakhir_pinjam': datetime.datetime(1111,1,11),
    'status' :'TERSEDIA'
}

data_perpustakaan = {}

while True:

  print(f"{"SELAMAT DATANG":^50}")
  print(f"{"SISTEM MANAJEMEN PERPUSTAKAAN":^50}")
  print("-"*50)

  perpustakaan = dict.fromkeys(perpustakaan_template.keys())
  #print(perpustakaan)

  perpustakaan['judul_buku'] = input("JUDUL BUKU :")
  perpustakaan['penulis'] = input("PENULIS :")
  perpustakaan['tahun_terbit'] = int(input("TAHUN TERBIT :"))

  print("\nMasukkan Tanggal Terakhir Pinjam :")
  TAHUN_PINJAM = int(input("Tahun pinjam (YYYY): "))
  BULAN_PINJAM = int(input("Bulan pinjam (1-12): "))
  TANGGAL_PINJAM = int(input("Tanggal pinjam (1-31): "))
  perpustakaan['tanggal_terakhir_pinjam'] = datetime.datetime(TAHUN_PINJAM,BULAN_PINJAM,TANGGAL_PINJAM)

  dipinjam = input("Apakah buku tersedia? (y/n): ").lower()
  if dipinjam == "y":
    perpustakaan["status"] = True
  else:
    perpustakaan["status"] = False

  KEY = ''.join((random.choice(string.ascii_uppercase)for i in range (6)))

  data_perpustakaan.update({KEY: perpustakaan})

  print("\n" + "-" * 120)
  print(f"{'KEY':<8} {'JUDUL BUKU':<25} {'PENULIS':<201} {'TAHUN TERBIT':<15} {'TANGGAL TERAKHIR PINJAM':<25} {'STATUS':<20}")
  print("-"*120)

  for lbry in data_perpustakaan:
    KEY = lbry

    JUDUL = data_perpustakaan[KEY]['judul_buku']
    PENULIS  = data_perpustakaan[KEY]['penulis']
    TERBIT = data_perpustakaan[KEY]['tahun_terbit']
    PINJAM = data_perpustakaan[KEY]['tanggal_terakhir_pinjam'].strftime("%x")
    STATUS = "Tersedia" if data_perpustakaan[KEY]['status'] else "Tidak"

    print(f"{KEY:<8} {JUDUL:<25} {PENULIS:<20} {TERBIT:<15} {PINJAM:<25} {STATUS:20}")
  print("\n")

  is_done = input("Mau tambah buku lagi (y/n)? ")
  if is_done == "n":
    break
```
